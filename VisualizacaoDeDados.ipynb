```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
valores_x = [1, 2, 4, 6, 8, 10, 12]
valores_y = [1, 2, 4, 6, 8, 10, 12]
valores_ny = [1, 2, 2, 4, 8, 16, 32]
```

```python
plt.plot(valores_x, valores_y, label='Valores', color='red')
plt.plot(valores_x, valores_ny, color='blue')
plt.xlim(1,12)
plt.ylim(0,32)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Gráfico Matplotlib') ;
```

**Tipos de Gráficos**

**Gráfico de dispersão**

```python
v_x = np.linspace(0, 20, 100)
v_y = np.cos(v_x)
```

```python
plt.plot(v_x, v_y, 'o') # o parâmetro 'o' indica que o gráfico é de dispersão
```

```python
# a função scatter também gera gráfico de disperssão
cores = np.random.rand(100)
plt.scatter(v_x, v_y, c=cores)
plt.colorbar() ;
```

```python
vb_x = np.arange(10)+1
vb_y = vb_x * 2

plt.bar(vb_x, vb_y) ;
```

**Gráfico de Barras Empilhadas**

```python
plt.bar(vb_x, vb_y, bottom=vb_y, width=0.8, color='gray')
plt.bar(vb_x, vb_y, bottom=vb_x*4, width=0.8, color='yellow')
```

**Gráficos de Barras Horizontais**

```python
plt.barh(vb_x, vb_y)
```

**Gráfico de Histograma**

```python
np.random.seed(15)
vh_x = np.random.normal(1,5,100)
plt.hist(vh_x)
```

**Gráfico de histograma em formato de matriz**

```python
df_tips = sns.load_dataset('tips')
df_tips.head(3)
```

```python
df_tips.time.value_counts()
```

```python
matriz = sns.FacetGrid(df_tips, row='sex', col='time', margin_titles=True, height=4)
matriz.map(plt.hist, 'tip', bins=10)
```

**Mapas de Calor (HeatMaps)**

```python
df_mpg = sns.load_dataset('mpg')
df_mpg.head(3)
```

```python
corr = df_mpg[df_mpg.columns[:-2]]
correlacoes = corr.corr()
correlacoes
```

```python
sns.heatmap(correlacoes)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlacoes, cmap='RdBu', annot=True)
```
